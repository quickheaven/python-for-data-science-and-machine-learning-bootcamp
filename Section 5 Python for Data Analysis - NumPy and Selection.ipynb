# NumPy Indexing and Selection

```python
import numpy as np
```

```python
arr = np.arange(0, 11)
```

```python
arr
```

```python
arr[8]
```

```python
arr[1:]
```

```python
arr[1:5]
```

```python
arr[0:5]
```

```python
arr[:6] # the same as saying 0:6
```

```python
arr[5:]
```

```python
arr[0:5] = 100 # broadcast in all 0:5
```

```python
arr
```

```python
arr = np.arange(0, 11)
```

```python
arr
```

```python
slice_of_arr = arr[0:6]
```

```python
slice_of_arr
```

```python
slice_of_arr[:] = 99
```

```python
slice_of_arr
```

```python
arr
```

```python
arr_copy = arr.copy()
```

```python
arr
```

```python
arr_copy[:] = 100
```

```python
arr_copy
```

```python
arr
```

## 2D Array

```python
arr_2d = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 45]])
```

```python
arr_2d
```

```python
# grabbing values in 2d array - double bracket and single bracket format (recommended)

# grab '5'
arr_2d[0][0]
```

```python
arr_2d[0]
```

```python
arr_2d[1][1]
```

```python
arr_2d[2][1]
```

```python
arr_2d[2, 1] # row comma column number
```

```python
arr_2d[1, 2]
```

```python
arr_2d[1][2]
```

```python
arr_2d[:2,1:] # slice notation - grab anything starting from row 2 and from column 1 onwards
```

```python
arr_2d[:2]
```

## Conditional Selections

```python
arr = np.arange(1, 11)
```

```python
arr
```

```python
arr > 5
```

```python
bool_arr = arr > 5
```

```python
bool_arr
```

```python
arr[bool_arr] # only returns instances where that boolean array is true
```

```python
arr[arr > 5] 
```

```python
arr[arr < 3]
```

```python
arr_2d = np.arange(50).reshape(5, 10)
```

```python
arr_2d
```

```python
arr_2d[1:3,]
```

```python
arr_2d[1:3,3:5]
```
