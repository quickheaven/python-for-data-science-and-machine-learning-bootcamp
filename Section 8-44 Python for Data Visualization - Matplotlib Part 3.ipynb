```python
import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x, y, color='orange')
#ax.plot(x, y, color='#FF8C00')

ax.plot(x, y, color='purple', linewidth=3, alpha=0.5) # linewidth=lw
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', linewidth=3, linestyle='--') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', linewidth=3, linestyle=':') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# linestyle supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
ax.plot(x, y, color='purple', linewidth=3, linestyle='dotted') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', linewidth=3, linestyle='-', marker='o', markersize=10) 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=0.5, ls='-', marker='+', markersize=20) 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=1, ls='-', marker='o', markersize=20) 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=1, ls='-', marker='o', markersize=20,
        markerfacecolor='yellow') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=1, ls='-', marker='o', markersize=20,
        markerfacecolor='yellow', markeredgewidth=3) 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=1, ls='-', marker='o', markersize=20,
        markerfacecolor='yellow', markeredgewidth=3,markeredgecolor='green') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=1, ls='--') 

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
