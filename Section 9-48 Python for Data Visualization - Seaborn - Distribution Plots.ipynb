```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=40)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips) # combine two distplot
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') # hex darker more points - less points lighter
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') # regression
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde') # kde kernel density estimation
```

```python
sns.pairplot(tips) # do jointplot for all combinations
```

```python
sns.pairplot(tips, hue='sex')  # hue - pass categorical
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm') # palette choose colors and styles
```

```python
sns.rugplot(tips['total_bill']) # draw a dash mark every single point
```
