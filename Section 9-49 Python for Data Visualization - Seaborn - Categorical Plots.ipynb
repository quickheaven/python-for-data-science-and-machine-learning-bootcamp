```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# barplot - general plot that allows to aggregate the categorical data. visuallization of group by action
sns.barplot(x='sex', y='total_bill', data=tips) 
```

```python
import numpy as np

sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std) 
```

```python
# countplot - is the same as bar plot except the estimator is explictily counting the number of occurences
# same as pandas count
sns.countplot(x='sex', data=tips) 
```

```python
# boxplot shows distribution of categorical data
# dots are outliers
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker') # hue - split further
```

```python
# unlike boxplot it allows to plot all datapoints
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
# draw a scatter plot with one variable is categorical
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
# combine a scatter plot / strip plot with a violin plot 
# drawback - does not scale on large number
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
# not for presentation
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```
