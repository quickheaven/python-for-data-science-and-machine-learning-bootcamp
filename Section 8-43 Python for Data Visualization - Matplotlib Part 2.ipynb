```python
import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y)
```

```python
fig, axes = plt.subplots() # almost the same as before but with subplots you can specify number of rows and columns

axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

#axes.plot(x, y)
plt.tight_layout() # fix overalpping, recommend to be added after your plot statements
```

```python
axes # axes is just a list, hence you loop
```

```python
for current_ax in axes:
    current_ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8, 2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))

axes[0].plot(x, y)

axes[1].plot(y, x)

plt.tight_layout()
```
